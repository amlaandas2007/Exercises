### Getting the data:

```python
test_path=r"D:\CODINGS\Machine_Learning\EXERCISE\Housing Price Competition\test.csv"
train_path=r"D:\CODINGS\Machine_Learning\EXERCISE\Housing Price Competition\train.csv"
```

```python
import pandas as pd

test_set = pd.read_csv(test_path)
train_set = pd.read_csv(train_path)
```

### Brief Info

```python
train_set.info()
```

### Selecting features

```python
features = train_set.select_dtypes(include = 'number').columns
features = features.drop(['Id','SalePrice'])
features
```

### Creating Test and Train Sets

```python
from sklearn.model_selection import train_test_split

X_full = train_set[features]
y_full = train_set['SalePrice']
X_test = test_set[features]
X_train, X_val, y_train, y_val = train_test_split(X_full, y_full, random_state = 1)
```

### Model Training

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

n_estimator_set = [50, 100, 150, 200, 250, 300, 350, 400]
n_estimator_select = 0
mae = None

for i in n_estimator_set:
    model_train = RandomForestRegressor(n_estimators = i, random_state = 1)
    model_train.fit(X_train, y_train)
    model_pred = model_train.predict(X_val)
    mae_cal = mean_absolute_error(y_val, model_pred)
    if mae is None:
        mae = mae_cal
    elif mae_cal < mae:
        n_estimator_select = i

best_n_estimator = n_estimator_select
```

### Selecting Best Model

```python
rf_reg = RandomForestRegressor(n_estimators= best_n_estimator)
rf_reg.fit(X_full, y_full)
```

### Making Pred:

```python
y_pred = rf_reg.predict(X_test)
y_pred
```

### Saving CSV file

```python
output = pd.DataFrame({'Id' : test_set.Id , 'SalePrice' : y_pred})
output.to_csv('submission.csv', index = False)
```
